## 4. Orders, payments, customers and reviews

### 4.1 Reviews data set

```python
olist_order_reviews.head()
```

```python
olist_order_reviews.order_id.is_unique,olist_order_reviews.review_id.is_unique
```

```python
olist_order_reviews.info()
```

We choose not to directly drop null rows for comment title, msg. Because it is very likely that a customer is not gonna write a comment. Furthermore, these rows represent almost half of the data set. We will just fill them with a "The customer did not comment".

```python
# Convering Dtype
olist_order_reviews = olist_order_reviews .astype({"review_creation_date" : 'datetime64[ns]', "review_answer_timestamp":'datetime64[ns]'})
olist_order_reviews.review_comment_message.fillna('The customer did not comment', inplace=True)
olist_order_reviews.review_comment_title.fillna('The customer did not comment', inplace=True)
```

```python
# Discard duplicates
olist_order_reviews.drop_duplicates(subset='order_id', inplace=True)
olist_order_reviews.drop_duplicates(subset='review_id', inplace=True)
olist_order_reviews.shape, olist_order_reviews.order_id.is_unique, olist_order_reviews.review_id.is_unique
```

### 4.2 Merging Orders, payments, customers and reviews

```python
ordersPaymentsCustomersReview = orders_customers_payments.merge(olist_order_reviews, on='order_id', how='inner')
```
